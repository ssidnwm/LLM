## 텍스트 추출 및 확인

```python
# 전체

with open(data_path, 'rb') as file:
    # Create a PdfReader object
    reader = PyPDF2.PdfReader(file)
    
    # Get the number of pages in the PDF
    #num_pages = len(reader.pages)
    #print(f'Total pages: {num_pages}')
    
    num_pages = 17
    
    text_sum = ""
    for page_num in range(3,num_pages):
        page = reader.pages[page_num]
        
        text_tmp = page.extract_text()
        lines = text_tmp.splitlines(True)
        
        lines = lines[3:]
        lines[0] = lines[0][2:]
        lines.append("\n")
        lines.append("[END OF PAGE]\n")
        
        text_tmp = ''.join(lines)
        text_sum = text_sum + text_tmp 
    
    # Print the text (or save it to a variable/file)
    print(text_sum)
```
